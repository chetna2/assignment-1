```python
centers = [[1, 1], [-1, -1], [1, -1]]
iris = datasets.load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
colors = {'Setosa':'#FCEE0C','Versicolor':'#FC8E72','Virginica':'#FC3DC9'}
fig = plt.figure(1, figsize=(8, 6))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=46, azim=130)

plt.cla()
pca = decomposition.PCA(n_components=3)
pca.fit(X)
X = pca.transform(X)

for color, i, target_name in zip(colors.values(), [0, 1, 2], target_names):
        ax.scatter(X[y==i,0], X[y==i,1], X[y==i,2], color=color,alpha=.8,cmap=plt.cm.spring,
           edgecolor='k',s=130)
        ax.legend(loc='best', shadow=False, scatterpoints=1)
    # Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(float)
#ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.spring,
           #edgecolor='k',depthshade='False')

plt.xlabel('1st PCA component')
ax.w_xaxis.set_ticklabels([])
plt.ylabel('2nd PCA component')
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd PCA component")
ax.w_zaxis.set_ticklabels([])
plt.title('3D of IRIS dataset')
plt.savefig('3D_Plot.jpg')
plt.show()





```
